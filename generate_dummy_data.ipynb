```python
import pandas as pd
import numpy as np
```

```python
legs_raw = pd.read_csv("leg_data.csv")
race_data = legs_raw[["Leg", "Leg Distance", "Rank", "Category"]]
race_data.head(5)
```

```python
runners = pd.read_csv("runner_data.csv")
ILRunners = runners[runners["Team"] == "Inner Loop"]
LTRunners = runners[runners["Team"] == "Loop Troop"]
RRunners = runners[runners["Team"] == "Radar"]
```

```python
def generate_race_time(distance_miles):
    # Generate a random pace between 7:30 and 8:30 per mile (in minutes)
    np.random.seed(np.random.randint(1000))
    pace_seconds = np.random.randint(450, 510)  # 7:30 min = 450 sec, 8:30 min = 510 sec
    
    # Calculate total time in seconds
    total_time_seconds = pace_seconds * distance_miles
    
    # Convert seconds to hours, minutes, and seconds
    hours = total_time_seconds // 3600
    minutes = (total_time_seconds % 3600) // 60
    seconds = total_time_seconds % 60
    
    return f"{int(hours):02d}:{int(minutes):02d}:{int(seconds):02d}"
```

```python
def generate_dummy_race_data(team):
    # Generate dummy data for testing
    np.random.seed(np.random.randint(50))
    race_data[team+" Time"] = legs_raw["Leg Distance"].apply(generate_race_time)
    
    # Add in the runenrs
    r = runners[runners["Team"] == team]
    race_data[team+" runner"] = pd.concat([r, r, r], ignore_index=True)["Name"]
```

```python
generate_dummy_race_data("Inner Loop")
generate_dummy_race_data("Loop Troop")
generate_dummy_race_data("Radar")
```
